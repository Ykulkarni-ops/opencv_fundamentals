```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# poiting the harcascade classifier to the xml file

face_classifier = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')

# loading the image and converting it to gray scale
image = cv2.imread('/content/yash.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Classifier detects the ROI of the faces in the image

faces = face_classifier.detectMultiScale(gray, 1.3, 5)

# If no faces detected return empty tuple

if faces is ():
  print(" No faces found")

# Iterating through all the faces in the image and
# drawing a bbox around them

for (x,y,w,h) in faces:
  cv2.rectangle(image, (x,y), (x+w, y+h), (0,255,0), 5)

imshow('Face detection', image)
```

```python
# eye and face detection together

image = cv2.imread('/content/family.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# loading the classifier
eye_classifier = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
face_classifier = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')

# detecting the faces
faces = face_classifier.detectMultiScale(gray, 1.3, 5)
eyes = eye_classifier.detectMultiScale(gray, 1.3, 5)

# if no faces or eyes found return empty tuple
if faces is ():
  print('No faces found')

# iterating through all the faces and draw bbox
# around face and eyes

for (x,y,w,h) in faces:
  cv2.rectangle(image, (x,y), (x+w, y+h), (0,255,0), 5)

for (ex,ey,ew,eh) in eyes:
  cv2.rectangle(image, (ex,ey), (ex+ew, ey+eh), (0,0,255), 5)

imshow('Face and eye detection', image)
```
